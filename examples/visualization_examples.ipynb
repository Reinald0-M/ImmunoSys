# ImmunoSys Visualization Examples

This notebook demonstrates the publication-ready visualization capabilities of the ImmunoSys framework.

```python
import numpy as np
import sys
sys.path.insert(0, '..')

from immunosys.visualization import (
    TimeSeries2D,
    PhasePlane2D,
    Plot3D,
    Heatmap,
    NetworkViz,
    quick_plot
)

import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Time Series Plots

Visualizing temporal dynamics of immune cell populations.

```python
# Generate example data: oscillating immune cell populations
time = np.linspace(0, 100, 1000)
t_cells = 50 + 30 * np.sin(0.1 * time) * np.exp(-0.01 * time)
b_cells = 40 + 20 * np.cos(0.15 * time) * np.exp(-0.01 * time)
nk_cells = 30 + 15 * np.sin(0.08 * time + 1) * np.exp(-0.01 * time)

data = np.column_stack([t_cells, b_cells, nk_cells])

# Create publication-ready time series plot
plotter = TimeSeries2D(style='science')
fig, ax = plotter.plot(
    time,
    data,
    labels=['T Cells', 'B Cells', 'NK Cells'],
    xlabel='Time (days)',
    ylabel='Cell Population',
    title='Immune Cell Dynamics',
    colors=['#1f77b4', '#ff7f0e', '#2ca02c'],
    show_grid=True
)
plt.show()
```

## 2. Phase Plane Diagrams

Visualizing the relationship between two state variables.

```python
# Generate phase plane data (predator-prey like dynamics)
theta = np.linspace(0, 4*np.pi, 500)
r = np.linspace(1, 3, 500)
x = r * np.cos(theta)
y = r * np.sin(theta)

# Create phase plane plot
plotter = PhasePlane2D(style='nature')
fig, ax = plotter.plot(
    x, y,
    xlabel='Effector Cells',
    ylabel='Target Cells',
    title='Phase Plane: Effector-Target Dynamics',
    show_direction=True,
    color_by_time=True,
    equilibria=[(0, 0)]
)
plt.show()
```

## 3. Vector Field

Visualizing the direction field of a 2D dynamical system.

```python
# Define a simple system (Lotka-Volterra)
def dx_dt(x, y):
    return x * (1 - 0.1*y)

def dy_dt(x, y):
    return -y * (1 - 0.1*x)

# Create vector field plot
plotter = PhasePlane2D()
fig, ax = plotter.plot_vector_field(
    x_range=(0, 20),
    y_range=(0, 20),
    dx_func=dx_dt,
    dy_func=dy_dt,
    density=20
)
plt.show()
```

## 4. 3D Trajectory Plots

Visualizing trajectories in 3D state space.

```python
# Generate 3D trajectory (Lorenz-like attractor)
t = np.linspace(0, 25, 2000)
x = np.sin(t) * np.cos(0.5*t)
y = np.sin(t) * np.sin(0.5*t)
z = np.cos(t)

# Create 3D trajectory plot
plotter = Plot3D(style='science')
fig, ax = plotter.plot_trajectory(
    x, y, z,
    xlabel='Naive Cells',
    ylabel='Activated Cells',
    zlabel='Memory Cells',
    title='3D Trajectory in Cell State Space',
    color_by_time=True
)
plt.show()
```

## 5. 3D Surface Plots

Visualizing parameter landscapes or fitness surfaces.

```python
# Generate surface data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2)) / (np.sqrt(X**2 + Y**2) + 0.1)

# Create 3D surface plot
plotter = Plot3D()
fig, ax = plotter.plot_surface(
    X, Y, Z,
    xlabel='Parameter α',
    ylabel='Parameter β',
    zlabel='Fitness',
    title='Parameter Fitness Landscape',
    cmap='plasma',
    alpha=0.9
)
plt.show()
```

## 6. Heatmaps

Visualizing spatial patterns or parameter sensitivities.

```python
# Generate heatmap data (spatial distribution)
np.random.seed(42)
spatial_data = np.random.randn(20, 30).cumsum(axis=0).cumsum(axis=1)

# Create heatmap
plotter = Heatmap(style='nature')
fig, ax = plotter.plot(
    spatial_data,
    xlabel='Spatial X',
    ylabel='Spatial Y',
    title='Spatial Distribution of Immune Cells',
    cmap='RdYlBu_r',
    cbar_label='Cell Density'
)
plt.show()
```

## 7. Parameter Sensitivity Analysis

Visualizing how different parameters affect model outputs.

```python
# Generate sensitivity data
parameters = ['α', 'β', 'γ', 'δ', 'ε']
outputs = ['Peak Time', 'Peak Value', 'Decay Rate']
sensitivity_matrix = np.random.randn(3, 5) * 2

# Create sensitivity heatmap
plotter = Heatmap()
fig, ax = plotter.plot_sensitivity(
    parameters,
    outputs,
    sensitivity_matrix,
    figsize=(10, 6)
)
plt.show()
```

## 8. Network Visualization

Visualizing cell-cell interaction networks.

```python
# Create example interaction network
n_cells = 6
adjacency = np.zeros((n_cells, n_cells))
# Add some interactions
adjacency[0, 1] = 1  # T cell -> B cell
adjacency[0, 2] = 1  # T cell -> NK cell
adjacency[1, 3] = 1  # B cell -> Antibody
adjacency[2, 4] = 1  # NK cell -> Target
adjacency[3, 5] = 1  # Antibody -> Target
adjacency[4, 0] = 1  # Target -> T cell (feedback)

cell_labels = ['T Cell', 'B Cell', 'NK Cell', 'Antibody', 'Target', 'Cytokine']
cell_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']

# Create network plot
plotter = NetworkViz(style='nature')
fig, ax = plotter.plot_network(
    adjacency,
    node_labels=cell_labels,
    node_colors=cell_colors,
    title='Immune Cell Interaction Network',
    layout='spring'
)
plt.show()
```

## 9. Quick Plotting

Using the convenience function for rapid visualization.

```python
# Quick line plot
data = np.random.randn(100, 3).cumsum(axis=0)
fig, ax = quick_plot(data, plot_type='line', labels=['A', 'B', 'C'])
plt.show()
```
